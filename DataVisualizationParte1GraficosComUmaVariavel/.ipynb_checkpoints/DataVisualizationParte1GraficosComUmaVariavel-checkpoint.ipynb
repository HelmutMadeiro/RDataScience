```r
dados  <-  read.csv(file = path, stringsAsFactors = FALSE)
str(dados)
```

```r
install.packages("dplyr")
install.packages("ggplot2")
```

```r
library(dplyr)
library(ggplot2)
```

```r
#tabela de frequencia sobre a coluna Rating
table(dados$Rating)

```

```r
#ploctando uma tabela de frequecia  com o hist
#limitando a linha x de 1 até 5 
hist(dados$Rating, xlim = c(1,5))
```

```r
#usando o ggplot 2
ggplot(data = dados) + # passando a base de dados e concatenando as demais instruções 
    geom_histogram(mapping = # aqui vai receber os eixos  
                             aes #realiza o mapeamentos dos valores que serão plotados 
                                (x = Rating),na.rm = TRUE # removendo valores no infinit 
                   ,breaks = seq(1,5)) + # diminir os valores de bins
                    xlim(c(1,5) # passando o limite
                   )

#alocando o resultado em uma variavel
rating.histogram  <- ggplot(data = dados) + geom_histogram(mapping = aes(x = Rating),na.rm = TRUE,breaks = seq(1,5)) +xlim(c(1,5))
```

```r
ggplot(data = dados) + 
    geom_bar (mapping = aes(x = Category)
              , stat = 'count' #Esse parâmetro indica se a própria função irá fazer a contagem de 
                               #cada valor semelhante à tabela de frequência
             )+ coord_flip()   #aqui tando o eixo x quanto o eixo y serão invertidos
```

```r
category.Freq  <- data.frame(table(dados$Category))
category.Freq
```

```r
ggplot(data = category.Freq ) + 
    geom_bar(mapping =aes(x= reorder(Var1,Freq) # ordernar decrescente ( Var1,-Freq) contrario)
                         ,y= Freq)
            ,stat = 'identity' # informa que nao é preciso passar pooistemos o y
            )+  
    coord_flip() 
```

```r
category.Top10 <-  category.Freq[(order(-category.Freq$Freq)), ] 
```

```r
category.Top10  <- category.Top10[1:10, ]
```

```r
ggplot(data = category.Top10) + 
    geom_bar(mapping =aes(x= reorder(Var1,Freq),y= Freq),stat = 'identity') + 
    coord_flip() 
```
